```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
def add_nodes(n):
    G=nx.Graph()
    G.add_nodes_from(range(n))
    return G    
G = add_nodes(20)
pos = nx.circular_layout(G)
nx.draw(G,pos,with_labels=1)
plt.show()
```

```python
def add_rand_edge(G):   
    i=np.random.choice(G.nodes())
    j=np.random.choice(G.nodes())
    if i!=j:
        G.add_edge(i,j,weight=random.randint(0,2))
    return G 
```

```python
def add_till_conn(G):
    while(nx.is_connected(G)==False):
        G = add_rand_edge(G)
    return G   
```

```python
G = add_till_conn(G)
print(nx.is_connected(G))
weight = nx.get_edge_attributes(G,'weight')
pos = nx.circular_layout(G)
nx.draw(G,pos,with_labels=1)
nx.draw_networkx_edge_labels(G,pos,edge_labels = weight)
plt.show()
```

```python
G.edges(data=True)
```

```python
cc = nx.closeness_centrality(G)
bc = nx.betweenness_centrality(G) 
dc = nx.degree_centrality(G) 
print("\t\t\tCentrality without considering weights\n")
print("Node \t Closeness Centrality \t Betweeness Centrality \t Degree Centrality:")
for i in range(0,20):
    print(i,"\t\t",round(cc[i],3),"\t\t\t",round(bc[i],3),"\t\t\t",round(dc[i],3))
```

```python
c = [(value, key) for key, value in cc.items()]
b = [(value, key) for key, value in bc.items()]
d = [(value, key) for key, value in dc.items()]
cm = max(c)
bm = max(b)
dm = max(d)
print("Max closeness centrality node is node",cm[1],"with value:",cm[0]) 
print("Max betweeness centrality node is node",bm[1],"with value:",bm[0]) 
print("Max degree centrality node is node",dm[1],"with value:",dm[0]) 
```

```python
cc = nx.closeness_centrality(G,distance = 'weight')
bc = nx.betweenness_centrality(G,weight = 'weight') 
dc = nx.degree_centrality(G) 
print("\t\t\tCentrality considering weights\n")
print("Node \t Closeness Centrality \t Betweeness Centrality \t Degree Centrality:")
for i in range(0,20):
    print(i,"\t\t",round(cc[i],3),"\t\t\t",round(bc[i],3),"\t\t\t",round(dc[i],3))
```

```python
c = [(value, key) for key, value in cc.items()]
b = [(value, key) for key, value in bc.items()]
d = [(value, key) for key, value in dc.items()]
cm = max(c)
bm = max(b)
dm = max(d)
print("Max closeness centrality node is node",cm[1],"with value:",cm[0]) 
print("Max betweeness centrality node is node",bm[1],"with value:",bm[0]) 
print("Max degree centrality node is node",dm[1],"with value:",dm[0]) 
```

# Conclusion 

A vertex with a high closeness centrality would mean it has close relationships with many vertices. Farness centrality is the reciprocal of the closeness centrality so that if the closeness is small, then the farness is large and vice versa.

For every pair of vertices in a connected graph, there exists at least one shortest path between the vertices such that either the number of edges that the path passes through (for unweighted graphs) or the sum of the weights of the edges (for weighted graphs) is minimized. The betweenness centrality for each vertex is the number of these shortest paths that pass through the vertex.

The degree centrality of a node is simply its degreethe number of edges it has. The higher the degree, the more central the node is. 

As we can see node 17 has highest closeness and degree centrality whereas node 2 has highest betweeness centrality when we are not considering weights.

As we can see node 17 has highest closeness,betweeness and degree centrality when we are considering weights.
