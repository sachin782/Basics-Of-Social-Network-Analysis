# Graphs.

```python
import networkx as nx
from networkx.algorithms import bipartite
import matplotlib.pyplot as plt
```

Undirected Graph.

```python
G = nx.Graph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
G.add_edge("A","B")
G.add_edge("B","C")
G.add_edge("C","E")
G.add_edge("C","F")
G.add_edge("D","E")
G.add_edge("F","G")
print("Nodes:",G.nodes())
print("Edges:",G.edges())
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, edge_color='r', arrows = True)
plt.show()
```

Undirected weighted graph.

```python
G = nx.Graph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
G.add_edge("A","B", weight = 6)
G.add_edge("B","C", weight = 2)
G.add_edge("C","E", weight = 3)
G.add_edge("C","F", weight = 1)
G.add_edge("D","E", weight = 8)
G.add_edge("F","G", weight = 3)
print("Nodes:",G.nodes())
print("Edges:",G.edges())
p = {"A":[-1,1],"B":[1,1],"C":[3,1],"D":[5,1],"E":[7,3],"F":[3,4],"G":[4,6]}
labels = nx.get_edge_attributes(G,'weight')
nx.draw(G,pos = p, with_labels=1)
nx.draw_networkx_edge_labels(G,pos = p,edge_labels=labels)
plt.show()
```

Relation in graph.

```python
G = nx.Graph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
r = [1, 2, 2, 4, 3, 1]
G.add_edge("A","B", relation = r[0])
G.add_edge("B","C", relation = r[1])
G.add_edge("C","E", relation = r[2])
G.add_edge("C","F", relation = r[3])
G.add_edge("D","E", relation = r[4])
G.add_edge("F","G", relation = r[5])
print("Nodes",G.nodes())
print("Edges",G.edges())
p = {"A":[-1,1],"B":[1,1],"C":[3,1],"D":[5,1],"E":[7,3],"F":[3,4],"G":[4,6]}
nx.draw(G, pos = p, with_labels = 1)
nx.draw_networkx_edges(G, pos = p, edge_color=r)
plt.show()
```

Roles in Graph.

```python
G = nx.Graph()
rel = [1, 2, 2, 4, 3, 1]
rol = [13, 11, 13, 11, 12, 10, 12]
G.add_node("A",role = rol[0])
G.add_node("B",role = rol[1])
G.add_node("C",role = rol[2])
G.add_node("D",role = rol[3])
G.add_node("E",role = rol[4])
G.add_node("F",role = rol[5])
G.add_node("G",role = rol[6])
G.add_edge("A","B", relation = rel[0])
G.add_edge("B","C", relation = rel[1])
G.add_edge("C","E", relation = rel[2])
G.add_edge("C","F", relation = rel[3])
G.add_edge("D","E", relation = rel[4])
G.add_edge("F","G", relation = rel[5])
print("Nodes",G.nodes())
print("Edges",G.edges())
p = {"A":[-1,1],"B":[1,1],"C":[3,1],"D":[5,1],"E":[7,3],"F":[3,4],"G":[4,6]}
nx.draw(G, pos = p, with_labels = 1)
nx.draw_networkx_edges(G, pos = p, edge_color = rel)
nx.draw_networkx_nodes(G, pos = p, node_color = rol)
plt.show()
```

```python
G = nx.MultiGraph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
r = [1, 2, 2, 4, 3, 1, 4, 3, 2, 4]
G.add_edge("A","B", relation = r[0])
G.add_edge("B","A", relation = r[1])
G.add_edge("B","C", relation = r[2])
G.add_edge("C","B", relation = r[3])
G.add_edge("C","E", relation = r[4])
G.add_edge("C","F", relation = r[5])
G.add_edge("F","C", relation = r[6])
G.add_edge("D","E", relation = r[7])
G.add_edge("E","D", relation = r[8])
G.add_edge("F","G", relation = r[9])
print("Nodes:",G.nodes())
print("Edges:",G.edges())
p = {"A":[-1,1],"B":[1,1],"C":[3,1],"D":[5,1],"E":[7,3],"F":[3,4],"G":[4,6]}
p1 = {"A":[-1.1,1.1],"B":[-1.1,1.1],"C":[3.1,1.1],"D":[5.1,1.1],"E":[7.2,3.2],"F":[3.1,4.1],"G":[4.1,6.1]}
nx.draw(G, pos = p,with_labels = True)
nx.draw_networkx_edges(G, pos = p)
nx.draw_networkx_edges(G, pos = p1)
plt.show()
```

Signed graph.

```python
G = nx.Graph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
G.add_edge("A","B", sign = '+')
G.add_edge("B","C", sign = '-')
G.add_edge("C","E", sign = '+')
G.add_edge("F","C", sign = '+')
G.add_edge("D","E", sign = '-')
G.add_edge("F","G", sign = '-')
print("Nodes:",G.nodes())
print("Edges",G.edges())
p = {"A":[-1,1],"B":[1,1],"C":[3,1],"D":[5,1],"E":[7,3],"F":[3,4],"G":[4,6]}
labels = nx.get_edge_attributes(G,'sign')
nx.draw(G,pos = p, with_labels=1)
nx.draw_networkx_edge_labels(G,pos = p,edge_labels=labels)
plt.show()
```

Complete Graph.

```python
H = nx.complete_graph(10)
print("Edges:",H.edges())
print("Nodes:",H.nodes())
nx.draw(H,with_labels=1)
plt.show()
```

Bipartite Graph.

```python
G = nx.DiGraph()
G.add_nodes_from(['A', 'B', 'C', 'D', 'E'], bipartite = 0)
G.add_nodes_from(['1', '2', '3', '4'], bipartite = 1)
G.add_edges_from([('A','2'),('B','1'),('C','4'),('D','3'),('E','2')])
r1 = [1, 1, 1, 1, 1, 2, 2, 2, 2]
r2 = [1, 2, 3, 4, 5]
print("Nodes:",G.nodes())
print("Edges:",G.edges())
pos = {"A":[1,9],"B":[1,7],"C":[1,5],"D":[1,3],"E":[1,1], "1":[4,9],"2":[4,7],"3":[4,5],"4":[4,3]}
nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_nodes(G, pos, node_color = r1)
nx.draw_networkx_edges(G, pos, edge_color = r2)
nx.draw_networkx_labels(G, pos)
plt.show()
```

Projected graph.

```python
G = nx.Graph()
G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'], bipartite = 0)
G.add_nodes_from(['1', '2', '3', '4'], bipartite = 1)
G.add_edges_from([('A','1'),('B','1'),('C','1'),('D','1'),('H','1'),('B','2'),('C','2'),('D','2'),('E','2'),('G','2'),('E','3'),('F','3'),('H','3'),('J','3'),('E','4'),('I','4'),('J','4')])
X = set(['1', '2', '3', '4'])
print("Nodes:",G.nodes())
print("Edges:",G.edges())
print("Sets:",bipartite.sets(G))
print("Is X bipartition:",bipartite.is_bipartite_node_set(G,X))
P = bipartite.projected_graph(G,X)
pos = nx.spring_layout(P)
nx.draw_networkx_nodes(P, pos)
nx.draw_networkx_edges(P, pos)
nx.draw_networkx_labels(P, pos)
plt.show()
```
