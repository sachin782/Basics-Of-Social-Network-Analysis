# Triadic Closure

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def triadic_closure(G):
    for u in G.nodes():
        for v in G.nodes():
            if u != v:
                if len(sorted(nx.common_neighbors(G,u,v))) > 0:
                    if G.has_edge(u,v) == False:
                        G.add_edge(u,v)
                        print("After adding edge between ",u," and ",v,":\n")
                        pos = nx.circular_layout(G)
                        nx.draw(G,pos,with_labels = True)
                        plt.show()
```

```python
G = nx.Graph()
G.add_edge(1,3)
G.add_edge(2,3)
edges = G.edges()
pos = nx.circular_layout(G)
nx.draw(G,pos,with_labels=True)
plt.show()
```

```python
triadic_closure(G)
```

```python
G = nx.Graph()
G.add_edge(1,3)
G.add_edge(1,5)
G.add_edge(2,3)
G.add_edge(2,5)
G.add_edge(3,4)
pos = nx.circular_layout(G)
nx.draw(G,pos,with_labels=True)
plt.show()
```

```python
triadic_closure(G)
```
