# Weak and Strong Ties

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = nx.Graph()
for i in range(1,5):
    G.add_edge('A',i)
    G.add_edge('B',i)
G.add_edge('A',7)
G.add_edge('A',8)
G.add_edge('B',9)
pos = nx.spring_layout(G)
nx.draw(G,pos,with_labels=1)
plt.show()
```

```python
def common_neighbor(G,A,B):
    c = 0
    for u in G.nodes():
        if G.has_edge(A,u) and G.has_edge(B,u):
            c = c + 1
    return c
```

```python
c = common_neighbor(G,'A','B')
print("Common neighbors =",c)
print("Ratio =",c,"/",len(G.nodes())-2)
```

```python
if(c >= 4):
    print("Strong Tie between A and B")
else:
    print("Weak Tie between A and B")
```
