# 8. Spatial Segregation

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

Creating a 10 X 10 matrix graph.

```python
N=10
G=nx.grid_2d_graph(N,N)
nx.draw(G)
plt.show()
```

```python
G.nodes()
```

Display matrix graph with positions.

```python
pos = dict((n,n) for n in G.nodes())
nx.draw(G,pos)
plt.show()
```

Display matrix graph with labels.

```python
labels=dict(((i,j), i*10+j) for i, j in G.nodes())
nx.draw(G,pos,with_labels=False)
nx.draw_networkx_labels(G,pos,labels=labels)
plt.show()
```

Right slant diagonal edges.

```python
for (u,v) in G.nodes():
    if ((u+1 <= N-1) and (v+1 <= N-1)):
        G.add_edge((u,v),(u+1,v+1))
nx.draw(G,pos,with_labels=False)
nx.draw_networkx_labels(G,pos,labels=labels)
plt.show()
```

Left slant diagonal edges.

```python
for (u,v) in G.nodes():
    if ((u+1 <= N-1) and (v-1 >=0)):
        G.add_edge((u,v),(u+1,v-1))
nx.draw(G,pos,with_labels=False)
nx.draw_networkx_labels(G,pos,labels=labels)
plt.show()
```

Randomly select type for each node.

```python
for n in G.nodes():
    G.node[n]['type']=random.randint(0,2)
for (n,d) in G.nodes(data=True):
    print(n,d)
```

Print nodes of type 1.

```python
type1_node_list=[]
for (n,d) in G.nodes(data=True):
    if d['type']==1:
        type1_node_list.append(n)
print("Type 1 nodes:")
print(type1_node_list)
label1=dict(((i,j), i*10+j) for i, j in type1_node_list)
nx.draw_networkx_nodes(G,pos,node_color='green',nodelist=type1_node_list)
nx.draw_networkx_labels(G,pos,labels=label1)
nx.draw_networkx_edges(G,pos)
plt.show()
```

Print nodes of type 2.

```python
type2_node_list=[]
for (n,d) in G.nodes(data=True):
    if d['type']==2:
        type2_node_list.append(n)
print("Type 2 nodes:")
print(type2_node_list)
label2=dict(((i,j), i*10+j) for i, j in type2_node_list)
nx.draw_networkx_nodes(G,pos,node_color='red',nodelist=type2_node_list)
nx.draw_networkx_labels(G,pos,labels=label2)
nx.draw_networkx_edges(G,pos)
plt.show()
```

Print nodes of type 0.

Print boundary nodes.

```python
def get_boundary_nodes(G):
    boundary_nodes_list=[]
    for (u,v) in G.nodes():
        if u==0 or u == N-1 or v==0 or v == N-1:
            boundary_nodes_list.append((u,v))           
    return boundary_nodes_list
boundary_nodes_list=get_boundary_nodes(G)
print("Boundary nodes:")
print(boundary_nodes_list)
```

Print internal nodes.

```python
internal_nodes_list=list(set(G.nodes())-set(boundary_nodes_list))
print("Internal nodes:")
print(internal_nodes_list)
```

Get neighbours for internal nodes.

```python
def get_neigh_for_internal(u,v):
    return [(u-1,v),(u+1,v),(u,v-1),(u,v+1),(u-1,v+1),(u+1,v-1),(u-1,v-1),(u+1,v+1)]
```

Get neighbours for boundary nodes.

```python
def get_neigh_for_boundary(u,v):
    if u==0 and v==0:
        return([(0,1),(1,0),(1,1)])
    elif u==N-1 and v==N-1:
        return([(u-1,v-1),(u,v-1),(u-1,v)])
    elif u==N-1 and v==0:
        return([(u-1,v),(u,v+1),(u-1,v+1)])
    elif u==0 and v==N-1:
        return([(u+1,v),(u+1,v-1),(u,v-1)])
    elif u==0:
        return([(u,v-1),(u,v+1),(u+1,v),(u+1,v-1),(u+1,v+1)])
    elif u==N-1:
        return([(u-1,v),(u,v-1),(u,v+1),(u-1,v+1),(u-1,v-1)])
    elif v==N-1:
        return([(u,v-1),(u-1,v),(u+1,v),(u-1,v-1),(u+1,v-1)])
    elif v==0:
        return([(u-1,v),(u+1,v),(u,v+1),(u-1,v+1),(u+1,v+1)])
```

Print unsatisfied nodes in the graph.

```python
def get_unsatisfied_nodes_list(G,internal_nodes_list,boundary_nodes_list):
    unsatisfied_nodes_list=[]
    threshold=3
    for (u,v) in G.nodes():
        type_of_this_node=G.node[(u,v)]['type']
        if type_of_this_node==0:
            continue
        else:
            similar_nodes=0
            if (u,v) in internal_nodes_list:
                neigh=get_neigh_for_internal(u,v)
            else:
                neigh=get_neigh_for_boundary(u,v)
        for each in neigh:
            if G.node[each]['type']==type_of_this_node:
                similar_nodes+=1
        if similar_nodes < threshold:
            unsatisfied_nodes_list.append((u,v))
    return unsatisfied_nodes_list
```

```python
unsatisfied_nodes_list=get_unsatisfied_nodes_list(G,internal_nodes_list,boundary_nodes_list)
print("Unsatisfied nodes:")
print(unsatisfied_nodes_list)
```

```python
satisfied_nodes_list=list(set(G.nodes())-set(unsatisfied_nodes_list)-set(empty_cells))
print("Satisfied nodes:")
print(sorted(satisfied_nodes_list))
```
